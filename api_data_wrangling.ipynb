```python
# Import Libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Optional: Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

## Understand the Open-Meteo API

- **API Endpoint**: `https://archive-api.open-meteo.com/v1/archive`
- **Parameters**:
  - `latitude`: Latitude of the location.
  - `longitude`: Longitude of the location.
  - `start_date`: Start date of the data (YYYY-MM-DD).
  - `end_date`: End date of the data (YYYY-MM-DD).
  - `hourly`: Comma-separated list of variables (e.g., temperature, precipitation).
  - `timezone`: Timezone of the data.

**Documentation**: [Open-Meteo API Docs](https://open-meteo.com/en/docs)

```python
# Step 1: Set Up Parameters

# Location coordinates for New York City
latitude = 40.7128
longitude = -74.0060

# Date range
start_date = '2023-01-01'
end_date = '2023-01-07'  # One week of data

# Variables to fetch
hourly_variables = ['temperature_2m', 'relativehumidity_2m', 'precipitation']
```

```python
# Step 2: Make the API Request

base_url = 'https://archive-api.open-meteo.com/v1/archive'

params = {
    'latitude': latitude,
    'longitude': longitude,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/New_York'
}

response = requests.get(base_url, params=params)
```

```python
# Check if the request was successful

if response.status_code == 200:
    print('Data fetched successfully!')
else:
    print(f'Failed to fetch data. Status code: {response.status_code}')
```

```python
# Step 3: Load Data into pandas DataFrame

data = response.json()
hourly_data = data['hourly']

df = pd.DataFrame(hourly_data)
df.head()
```

```python
# Step 4: Data Cleaning

# Convert 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])

# Set 'time' as the index
df.set_index('time', inplace=True)

# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if any)
df.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
# Step 5: Exploratory Data Analysis

# Summary Statistics
df.describe()
```

```python
# Plot Temperature Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['temperature_2m'], label='Temperature (°C)')
plt.title('Temperature Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
# Plot Relative Humidity Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['relativehumidity_2m'], color='orange', label='Relative Humidity (%)')
plt.title('Relative Humidity Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

```python
# Plot Precipitation Over Time

plt.figure(figsize=(14, 6))
plt.bar(df.index, df['precipitation'], label='Precipitation (mm)')
plt.title('Precipitation Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Step 6: Correlation Analysis

# Calculate correlation between temperature and humidity
correlation = df['temperature_2m'].corr(df['relativehumidity_2m'])
print(f'Correlation between Temperature and Relative Humidity: {correlation:.2f}')
```

```python
# Scatter Plot of Temperature vs. Relative Humidity

plt.figure(figsize=(8, 6))
plt.scatter(df['temperature_2m'], df['relativehumidity_2m'], alpha=0.5)
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
# Step 7: Resample Data to Daily Averages

daily_avg = df.resample('D').mean()
daily_avg.head()
```

```python
# Plot Daily Average Temperature

plt.figure(figsize=(10, 5))
plt.plot(daily_avg.index, daily_avg['temperature_2m'], marker='o')
plt.title('Daily Average Temperature in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
# Step 8: Compare with Another Location (Los Angeles)

# Coordinates for Los Angeles
latitude_la = 34.0522
longitude_la = -118.2437

params_la = {
    'latitude': latitude_la,
    'longitude': longitude_la,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/Los_Angeles'
}

response_la = requests.get(base_url, params=params_la)

# Load and clean Los Angeles data
data_la = response_la.json()
hourly_data_la = data_la['hourly']
df_la = pd.DataFrame(hourly_data_la)

df_la['time'] = pd.to_datetime(df_la['time'])
df_la.set_index('time', inplace=True)
df_la.fillna(method='ffill', inplace=True)
```

```python
# Resample Los Angeles Data to Daily Averages

daily_avg_la = df_la.resample('D').mean()
```

```python
# Combine DataFrames for Comparison

combined_temp = pd.DataFrame({
    'New York': daily_avg['temperature_2m'],
    'Los Angeles': daily_avg_la['temperature_2m']
})
```

```python
# Plot Comparison of Daily Average Temperature

combined_temp.plot(kind='bar', figsize=(10, 6))
plt.title('Daily Average Temperature: New York vs Los Angeles')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Step 9: Advanced Analysis - Temperature Difference

# Calculate Temperature Difference
combined_temp['Temp Difference'] = combined_temp['Los Angeles'] - combined_temp['New York']
combined_temp
```

```python
# Step 10: Save Data to CSV (Optional)

# Save New York data
df.to_csv('new_york_weather.csv')

# Save Los Angeles data
df_la.to_csv('los_angeles_weather.csv')
```

# **SUMMARY OF THE RESULT**
**Seattle** = Strong seasonal contrast: dry summers, extremely wet winters. Big weekly spikes (storm-driven rainfall).

**London** = More balanced: rain spread more evenly, smaller peaks, but almost no completely dry season.

**Overall**: Seattle had much more total rainfall and higher variability, while London had consistent but lighter rainfall.

```python
# STEP 1 — Set up parameters (London)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests

# London coordinates and timezone
latitude = 51.5074
longitude = -0.1278
timezone = "Europe/London"

# Date range (full year 2022)
start_date = "2022-01-01"
end_date   = "2022-12-31"

# Hourly variables to fetch (precipitation in mm)
hourly_variables = ["precipitation"]
```

```python
#STEP 2

base_url="https://archive-api.open-meteo.com/v1/archive"

params= {
    'latitude': latitude,
    'longitude': longitude,
    'start_date': start_date,
    'end_date': end_date,
    'hourly' : ','.join(hourly_variables),
    'timezone': timezone,
}
#send the request
response=requests=requests.get(base_url, params=params)
# Check if the request was successful
if response.status_code == 200:
    print("✅ Data fetched successfully!")
    print("Request URL:", response.url)
else:
    print(f"❌ Failed to fetch data. Status code: {response.status_code}")

```

```python
#STEP 3

data = response.json()
hourly_data = data['hourly']

df = pd.DataFrame(hourly_data)
df.head()

```

```python
#STEP 4
# data cleaning

df['time'] = pd.to_datetime(df['time'])
df.set_index('time', inplace=True)
df.isnull().sum()
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
#STEP 5

#summary
df.describe()
```

```python
plt.figure(figsize=(14,6))
plt.bar(df.index, df['precipitation'], label='Precipitation (mm)')
plt.title("Precipitation Over Time")
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
#STEP 6 skiped since we only look at precipitation

#STEP 7

daily_precip=df["precipitation"].resample("D").sum()

daily_precip.head()
```

```python
# Step 7b: Plot Daily Total Precipitation

plt.figure(figsize=(12,6))
plt.plot(daily_precip.index, daily_precip.values, marker='o', linestyle='-', label='Daily Precipitation (mm)')
plt.title("Daily Total Precipitation")
plt.xlabel("Date")
plt.ylabel("Precipitation (mm)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# STEP 8 — Add second city (Seattle), fetch & clean

import requests  # yeniden import ediyoruz, requests bozulmuş olabilir

# Seattle coordinates
latitude_sea = 47.6062
longitude_sea = -122.3321
timezone_sea = "America/Los_Angeles"

# API base URL
base_url = "https://archive-api.open-meteo.com/v1/archive"

# Parameters for Seattle
params_sea = {
    "latitude": latitude_sea,
    "longitude": longitude_sea,
    "start_date": start_date,
    "end_date": end_date,
    "hourly": ",".join(hourly_variables),
    "timezone": timezone_sea,
}

# Send request
response_sea = requests.get(base_url, params=params_sea)

# Parse JSON
data_sea = response_sea.json()
hourly_sea = data_sea["hourly"]

# Convert to DataFrame
df_sea = pd.DataFrame(hourly_sea)

# Cleaning
df_sea["time"] = pd.to_datetime(df_sea["time"])
df_sea.set_index("time", inplace=True)
df_sea.fillna(method="ffill", inplace=True)

# Quick check
df_sea.head()
```

```python
# STEP 8 b — Compare London vs Seattle (WEEKLY totals)

# 1) Resample hourly precipitation to weekly totals (mm/week)
weekly_london  = df["precipitation"].resample("W-SUN").sum()
weekly_seattle = df_sea["precipitation"].resample("W-SUN").sum()

combined_weekly = pd.concat([weekly_london, weekly_seattle], axis=1)
combined_weekly.columns = ["London (mm)", "Seattle (mm)"]

plt.figure(figsize=(12,6))
plt.plot(combined_weekly.index, combined_weekly["London (mm)"],  label="London (mm)")
plt.plot(combined_weekly.index, combined_weekly["Seattle (mm)"], label="Seattle (mm)")
plt.title("Weekly Total Precipitation — London vs Seattle (2022)")
plt.xlabel("Week")
plt.ylabel("Precipitation (mm)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

combined_weekly.head()
```

```python
# STEP 9 — Annual totals and difference

# 1) Calculate total annual precipitation for each city
total_london  = weekly_london.sum()
total_seattle = weekly_seattle.sum()

# 2) Calculate the difference (Seattle - London)
difference = total_seattle - total_london

# 3) Print results
print("Total Precipitation in London (2022):", round(total_london, 2), "mm")
print("Total Precipitation in Seattle (2022):", round(total_seattle, 2), "mm")
print("Difference (Seattle - London):", round(difference, 2), "mm")
```
